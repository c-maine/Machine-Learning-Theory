```python
import numpy as np
import math
import matplotlib.pyplot as plt
from numpy import (array, dot, arccos, clip)
from numpy.linalg import norm
```

```python
# Function to create random vectors over the unit ball

def rando_ball(d):
    N = np.random.normal(0,1,d)
    N_norm = np.linalg.norm(N)
    N = N/N_norm
    return(N)
```

```python
# Function to create random vectors on unit cube

def rando_cube(d):
    N = np.random.uniform(low=-1, high=1, size=d)
    return(N)
```

```python
# Function to calculate pairwise angles in radians

def angle_between(v1, v2):
    v1= np.array(v1)
    v2= np.array(v2).T
    dot_pr = v1.dot(v2)
    norms = np.linalg.norm(v1) * np.linalg.norm(v2)
    rez = np.rad2deg(np.arccos(dot_pr / norms))
    return(math.radians(rez))

# abs(angle_between(rando_cube(100), rando_cube(100))-1.57)
```

```python
# Function to calculate number of independent vectors you can add before one ceases to be nearly orthogonal

def iterator(method, d, epsilon, max_iters):
    
    epsilon = math.radians(epsilon)
    orthogonal_vectors = [method(d)]
    res=1
        
    v_i = method(d).reshape(1,d)
    angle = angle_between(v_i, orthogonal_vectors[0])
    if abs(angle-1.5708) < epsilon:
        orthogonal_vectors= orthogonal_vectors + [v_i]
    
        for i in range(1, max_iters, 1):
            v_i = method(d).reshape(1,d)
            c = 0

            for j in orthogonal_vectors:
                angle = angle_between(v_i, j)
                if abs(angle - 1.5708) < epsilon:
                    c += 1
                else: pass
                
            if c == len(j):
                orthogonal_vectors= orthogonal_vectors + [v_i]
        res = len(orthogonal_vectors)
      
    return(res) 
```

```python
# Function to calculate max number of vectors you can add before one ceases to be nearly orthogonal in a # of runs

def max_orth_vectors(method, d, epsilon, max_iters, runs):
    max_orthogonal_vectors = 0
    for i in range(0,runs,1):
        maybe_max = iterator(method, d, epsilon, max_iters)
        if maybe_max > max_orthogonal_vectors:
            max_orthogonal_vectors = maybe_max
        else: pass
    return(max_orthogonal_vectors)
```

```python
# Function to iterate over values of epsilon

def e_iterator(method, d, epsilon_range, runs, max_iters=1000):
    res =[]
    for epsilon in epsilon_range:
        res.append(max_orth_vector(method, d, epsilon, max_iters, runs))
    return(res)   
```
