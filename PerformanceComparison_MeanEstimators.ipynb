# 5.2 Functions to calculate estimators

```python
# Trimmed mean estimator function

def tri_mean(X, k):
    X.sort()
    tri_X = X[k:(len(X)-k)]
    res = emp_mean(tri_X)
    return(res)
```

```python
# Empirical mean estimator function

def emp_mean(X):
    res= sum(X)/len(X)
    return(res)
```

```python
# Function to split vector into chunks

def baskets_from(items, maxbaskets=25):
    baskets = [[] for _ in range(maxbaskets)] # in Python 3 use range
    for i, item in enumerate(items):
        baskets[i % maxbaskets].append(item)
    return (baskets)
```

```python
# Median of means estimator function

def med_mean(X, k):
    splits = baskets_from(X, k)
    means = []
    for i in splits:
        mean=np.mean(i)
        means.append(mean)
    med_mean = np.median(means)
    return(med_mean)
```

# 5.3 Functions to generate iiD RVs

```python
# Generate n iid random variables from different distributions

# Light tails

def gauss_gen(n, miu=0, sigma=1):
    res = np.random.normal(miu, sigma, n)
    return(res)

def laplace_gen(n, mean=0, decay=1):
    res = np.random.laplace(loc=mean, scale=decay, size=n)
    return(res)

# Heavy tails

def pareto_gen(n):
    res=[]
    for i in range(0,n):
        rando_pareto = np.random.pareto(n)
        res.append(rando_pareto)
    return(res)
 
def t_gen(n, df, size=None):
    res=[]
    for i in range(0,n):
        rando_t = np.random.standard_t(df, size)
        res.append(rando_t)
    return(res)
    
```

# 5.4 Function to estimate P{|mn-m| > eps}

```python
# Function to estimate P{|mn-m| > }

def p_mn_e(X, m, epsilon):
    ms = [m] * len(X)
    resid = np.subtract(np.array(X),np.array(ms))
    resid = abs(resid)
    temp_res = (resid > epsilon).sum()
    res = temp_res / len(X)
    return(res)
```

# 5.5 Iterator functions

```python
# Function to calculate probability of |mn-m| >e over large sample

def iterator(num_iters, distrib, n, estimator, true_m, epsilon, tri_k, med_mea_k, t_df):

    estimators = []
    
    if distrib == t_gen:  
    
        if estimator == tri_mean:
            for i in range(0, num_iters):
                sample = t_gen(n, t_df)
                est = tri_mean(sample, tri_k)
                estimators.append(est)

        elif estimator == med_mean:
            for i in range(0, num_iters):
                sample = t_gen(n, t_df)
                est = med_mean(sample, med_mea_k)
                estimators.append(est)

        else:
            for i in range(0, num_iters):
                sample = t_gen(n, t_df)
                est = estimator(sample)
                estimators.append(est)
        
    
    else: 

        if estimator == tri_mean:
            for i in range(0, num_iters):
                sample = distrib(n)
                est = tri_mean(sample, tri_k)
                estimators.append(est)

        elif estimator == med_mean:
            for i in range(0, num_iters):
                sample = distrib(n)
                est = med_mean(sample, med_mea_k)
                estimators.append(est)

        else:
            for i in range(0, num_iters):
                sample = distrib(n)
                est = estimator(sample)
                estimators.append(est)
    
    return(p_mn_e(estimators, true_m, epsilon)) 

# Test
# iterator(1000, gauss_gen, 100, med_mean, 0, 0.3, 45,1)
```

```python
# Function to calculate probability of |mn-m| >e for different values of epsilon

def epsilon_iterator(num_iters, distrib, n, estimator, true_m, epsi_range, tri_k, med_mea_k, t_df):    
    epsi_iters = []    
    for i in epsi_range:
        prob_mn_m_e = iterator(num_iters, distrib, n, estimator, true_m, i, tri_k, med_mea_k, t_df)
        epsi_iters.append(prob_mn_m_e)
    return(epsi_iters)
```

```python
# Function to calculate probability of |mn-m| >e for different values of n

def n_iterator(num_iters, distrib, n_range, estimator, true_m, epsilon,tri_k, med_mea_k, t_df):    
    n_iters = []    
    for i in n_range:
        prob_mn_m_e = iterator(num_iters, distrib, i, estimator, true_m, epsilon, tri_k, med_mea_k, t_df)
        n_iters.append(prob_mn_m_e)
    return(n_iters)
```
